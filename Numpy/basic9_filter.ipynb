```python
# using query()
# query() - used to filter rows from database

# note -->
#  paramtr dhold contain valid string
#  col name shold not contain space
```

```python
import pandas as pd
df = pd.read_csv("emp.csv" )
df.head()
```

```python
df.query("eadrs == 'doon'")
```

```python
df.query('eadrs == "doon"')   # strng may create problm
```

```python
# not from doon

df.query("eadrs != 'doon'")
```

```python
# salary > 3000

# df.query("esal > 3000")

df[df["esal"]>3000]
```

```python
df
```

```python
df[(df["eadrs"] == "shimla") | (df["eadrs"] =="doon")]
```

```python
df[df["eadrs"].isin(["doon" , "shimla" , "noida"])]
```

```python
for i in df.index:
    if df["esal"][i]>4000:
        print(df.iloc[i:i+1,i:])
```

```python
# multiple conditions
```

```python
# doon emp salry > 3500
```

```python
df.query("eadrs == 'doon' and esal > 3500")
```

```python
# & , | 
```

```python
# not from doon or salary>5000
```

```python
df.query("eadrs != 'doon' or esal > 5000")
```

```python
# use of in , not in
```

```python
df.query("eadrs == ['doon' , 'delhi']")
```

```python
# or

df.query("eadrs in  ['delhi' , 'doon']")
```

```python
# if column contain space

df.columns = ["a" , "b" , "a c"]
```
